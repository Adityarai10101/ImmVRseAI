```python
import cv2
```

```python
vidcap = cv2.VideoCapture('16SecondVideo.mp4')
success,image = vidcap.read()
count = 0
success = True
while success:
  success,image = vidcap.read()
  success,image = vidcap.read()
  success,image = vidcap.read()
  success,image = vidcap.read()
  success,image = vidcap.read()
  success,image = vidcap.read()
  success,image = vidcap.read()
  cv2.imwrite("frameSet/frame%d.jpg" % count, image)     # save frame as JPEG file
  if cv2.waitKey(10) == 27: # exit if Escape is hit
      break
  count += 1
```

```python
# Importing Image class from PIL module
from PIL import Image
import os

# Opens a image in RGB mode
for imageName in os.listdir("bg-RemovedFrameSet"):
    print(imageName)
    im = Image.open(os.path.join("C:/Users/raia7/Documents/Hackathon/frames/bg-RemovedFrameSet", imageName))
    width, height = im.size
    # Setting the points for cropped image
    uPoint = 0
    left = 420 + uPoint
    top = uPoint
    right = 1920 - (420 + uPoint)
    bottom = height - uPoint
    im1 = im.crop((left, top, right, bottom)).resize((512,512))
    im1.save("512Transparent/{}".format(imageName))
```

```python
for imageName in os.listdir("512Transparent"):
    print(imageName)

    imgNew = Image.open(os.path.join("C:/Users/raia7/Documents/Hackathon/frames/512Transparent", imageName))
    imgNew = imgNew.convert("RGBA")
    datas = imgNew.getdata()
    newData = []
    for item in datas:
        if item[0] == 0 and item[1] == 0 and item[2] == 0 and item[3] == 0:
            newData.append((0, 0, 0, 255))
        else:
            newData.append((255,255,255,255))
    imgNew.putdata(newData)
    imageNameArr = imageName.split(".")
    imageName = imageNameArr[0] + "_mask." + imageNameArr[1]
    imgNew.save("512_Mask/{}".format(imageName), "PNG")
```
